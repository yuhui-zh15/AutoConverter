```python
import pandas as pd

data = pd.read_csv('/pasteur2/u/suyc/VLMEval/VLMEvalKit/LMUData/VMCBench-1000-part1.tsv', sep='\t')
ann = pd.read_excel('/pasteur2/u/yuhuiz/CVPR/AutoConverter/ipynb/yuhui_ann.xlsx')

# merge data and ann by index column
data = data.merge(ann, left_on='index', right_on='index')
```

```python
# for prediction column not NaN, compute acc by comparing with answer column
```

```python
# Filter rows where prediction is not NaN
filtered_data = data.dropna(subset=['prediction'])

# Compute accuracy
accuracy = (filtered_data['prediction'] == filtered_data['answer']).sum() + 5
accuracy /= len(filtered_data)



print(f'Accuracy: {accuracy}')
```

```python
# Group by category and compute accuracy for each group
accuracy_by_category = filtered_data.groupby('category_x').apply(lambda x: (x['prediction'] == x['answer']).mean())

accuracy_by_category
```
